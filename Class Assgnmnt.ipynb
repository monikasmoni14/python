```python
class OddEven():
    def oddEven():
        num=int(input("enter the number:"))
        if((num%2)==1):
            print("Even number:")
            verify="even number"
        else:
            print("odd number:")
            verify="odd number"
        return verify
```

```python
class triAngle():
    def triangle():
        height = float(input("Height:"))
        breadth = float(input("Breadth:"))
        print("Area formula:  (Height*Breadth)/2")
        area = (height * breadth) / 2
        print("Area of Triangle:",area)
```
